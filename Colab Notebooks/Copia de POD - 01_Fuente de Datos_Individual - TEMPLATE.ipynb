# Base de datos - Relacionales

SQLite es una biblioteca de C que provee una base de datos ligera basada en disco que no requiere un proceso de servidor separado y permite acceder a la base de datos usando una variación no estándar del lenguaje de consulta SQL. Algunas aplicaciones pueden usar SQLite para almacenamiento interno. También es posible prototipar una aplicación usando SQLite y luego transferir el código a una base de datos más grande como PostgreSQL u Oracle.

Realizamos el `import` e indicamos el archivo en el que vamos a guardar la base de datos:

```python
import sqlite3
conn = sqlite3.connect('db.db')
```

La siguiente linea de codigo agrega filas a la misma. Por favor, ejecutar la siguiente celda para la consigna:



```python
!wget https://people.sc.fsu.edu/~jburkardt/data/csv/snakes_count_10000.csv

cur = conn.cursor()
cur.execute("CREATE TABLE t (GameNumber, GameLength);")

with open('snakes_count_10000.csv','r') as csv_file:
  lines = csv_file.readlines()[1:]
  lines = [l.split(',') for l in lines]
  to_db = [(l[0].strip(), l[1].strip()) for l in lines]

cur.executemany("INSERT INTO t (GameNumber, GameLength) VALUES (?, ?);", to_db)
conn.commit()
conn.close()
```

## Consigna

Hacer una request GET e imprimir uno de los campos de la respuesta:

```python
import requests

url = "https://hp-api.herokuapp.com/api/characters"

response = requests.get(url)

if response.status_code == 200:
    characters = response.json()
    print(f"El nombre del primer personaje es: {characters[0]['name']}")

else:
    print(f"Error en la solicitud. Código de estado: {response.status_code}")
```

```python
url = "https://hp-api.herokuapp.com/api/characters"

response = requests.get(url)


if response.status_code == 200:

    characters = response.json()

    for character in characters:

        gender = character.get('gender', '').lower()

        if gender == 'female':
            character_type = "Mujer"
        elif gender == 'male':
            character_type = "Hombre"
        else:
            character_type = "Desconocido"
        character['Tipo de personaje'] = character_type

        print(f"Personaje: {character['name']}")
        print(f"Tipo de personaje: {character['Tipo de personaje']}")
        print("-----")
else:
    print(f"Error en la solicitud. Código de estado: {response.status_code}")
```

Hacer un plot con `matplotlib` de una variable de una API de datos streaming en el que el eje 'y' sea la variable y el eje 'x' el tiempo.

```python
pip install websocket-client matplotlib
```

```python
import websocket
import json
import matplotlib.pyplot as plt
import datetime

# Inicializamos las listas para guardar los datos
times = []
prices = []

# Función que maneja los mensajes recibidos
def on_message(ws, message):
    data = json.loads(message)

    # Imprimir todo el mensaje para depuración
    print(f"Mensaje recibido: {data}")

    # Solo procesamos el mensaje si contiene el tipo 'ticker' (es decir, el precio de BTC-USD)
    if data.get('type') == 'ticker':
        price = float(data['price'])  # Obtener el precio actual
        time = datetime.datetime.now()  # Obtener el tiempo actual

        # Agregar los datos a las listas
        times.append(time)
        prices.append(price)

        # Mostrar el último precio recibido
        print(f"Precio actualizado de BTC-USD: ${price} en {time}")

        # Graficar los datos en tiempo real
        plt.clf()  # Limpiar la figura para evitar que se acumule información
        plt.plot(times, prices)
        plt.xlabel('Tiempo')
        plt.ylabel('Precio de BTC-USD')
        plt.title('Precio de BTC-USD en Tiempo Real')
        plt.xticks(rotation=45)
        plt.tight_layout()  # Ajustar el diseño para evitar que se corten etiquetas
        plt.pause(1)  # Hacer una pausa de 1 segundo para actualizar el gráfico

# Función que maneja la apertura de la conexión WebSocket
def on_open(ws):
    # Suscribirse al canal "ticker" para el par BTC-USD
    params = {
        "type": "subscribe",
        "channels": [
            {
                "name": "ticker",
                "product_ids": ["BTC-USD"]
            }
        ]
    }
    # Enviar la solicitud de suscripción
    ws.send(json.dumps(params))

# Configuración del WebSocket para Coinbase Pro
ws = websocket.WebSocketApp(
    "wss://ws-feed.pro.coinbase.com",  # URL WebSocket de Coinbase Pro
    on_message=on_message,  # Función que maneja los mensajes recibidos
    on_open=on_open         # Función que maneja la apertura de la conexión
)

# Configuración de matplotlib para graficar en tiempo real
plt.ion()  # Activar el modo interactivo de matplotlib
plt.figure(figsize=(10, 6))

# Ejecutar la conexión WebSocket
ws.run_forever()

```
